```python
# Import the dependencies
import pandas as pd
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
import seaborn as sns
import random
%matplotlib inline
```

```python
df = pd.read_csv('../zippedData/tn.movie_budgets.csv.gz')
df
```

```python
df.info()
```

```python
# Remove dollar signs ($) and commas from the strings, then convert to float
df[['production_budget', 'domestic_gross', 'worldwide_gross']] = (
    df[['production_budget', 'domestic_gross', 'worldwide_gross']]
    .replace({'\$': '', ',': ''}, regex=True)
    .astype(float)
)

# Convert the 'release_date' column to datetime type
df['release_date'] = pd.to_datetime(df['release_date'], format='%b %d, %Y')
df.info()
```
